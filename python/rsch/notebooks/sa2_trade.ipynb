```python
from etl import load_secid_data

def load_trade_data():
    btc = load_secid_data(
        trading_type='um', 
        market_data_type='trades', 
        time_period='monthly', 
        symbol='BTCUSDT', 
        interval=None,
        time_period_filename='2020-01'
    )

    btc = btc.sort_values(by=['time', 'trade_id'], ascending=[True, True])
    btc = btc.set_index('trade_id', drop=False)

    return btc

# btc = load_trade_data()
btc = lib.read('BTCUSDT').data
```

```python
dfi['price_x_qty'] = dfi['price'] * dfi['qty']

    
vwap = (
    dfi.set_index('time')['price_x_qty'].resample('1T').sum() / \
    dfi.set_index('time')['qty'].resample('1T').sum()
)
```

```python
dfi['price_x_qty'] = dfi['price'] * dfi['qty']

dfi.set_index('time')[['price_x_qty', 'qty']].resample('1T').sum().assign(vwap=lambda x: x['price_x_qty'] / x['qty'])
```

```python
dfi_vwap = (
    dfi
    .groupby('minute')
    .agg({'price_x_qty': 'sum', 'qty': 'sum'})
    .assign(
        vwap=lambda x: x['price_x_qty'] / x['qty'],
    )
)

dfi_vwap
```

```python
dfi_vwap.vwap.plot()
```

```python
import numba
from numpy import ndarray


@numba.njit
def make_vwap(price: ndarray, qty: ndarray) -> float:
    return (price * qty).sum() / qty.sum()

# (df.price * df.qty).sum() / df.qty.sum()

dfi = btc[btc['time'] < pd.Timestamp('2020-01-01 01:00:00')].copy()
dfi['minute'] = dfi['time'].dt.floor('T')

dfi.groupby('minute')[['price', 'qty']].apply(lambda x: make_vwap(x['price'].values, x['qty'].values))
```
