```python
import re

def strip_styles(html: str) -> str:
    # Regex to find style attributes
    style_regex = re.compile(r'\s*style="[^"]*"', re.IGNORECASE)
    # Substitute the style attributes with an empty string
    cleaned_html = style_regex.sub('', html)
    return cleaned_html

# Example usage
html_content = '''
<div style="color: red; font-size: 14px;">This is a div</div>
<p style="margin: 10px;">This is a paragraph</p>
<span style="background-color: yellow;">This is a span</span>
'''

cleaned_html = strip_styles(html_content)
print(cleaned_html)
```
